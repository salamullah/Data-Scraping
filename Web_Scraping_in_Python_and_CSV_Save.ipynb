```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url= "https://islamqa.info/en/answers/11497/how-to-make-wudu"
```

```python
page= requests.get(url)
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
topic1= soup.find(attrs={'class': 'title is-4 is-size-5-touch'}).text
```

```python
topic1
```

```python
topic= soup.find(attrs={'class': 'title is-4 is-size-5-touch'}).text.replace("\n", "")
```
