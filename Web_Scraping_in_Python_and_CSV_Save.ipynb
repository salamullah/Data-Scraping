```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url= "https://islamqa.info/en/answers/1/interruption-of-wudhu"
```

```python
page= requests.get(url)
```

```python
page.content
```

```python
soup= BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
topic1= soup.find(attrs={'class': 'title is-4 is-size-5-touch'}).text
```

```python
topic1
```

```python
topic2= soup.find(attrs={'class': 'has-text-weight-bold'}).text
```

```python
topic2
```

## The Above Topic 2 is best

```python
topic= soup.find(attrs={'class': 'title is-4 is-size-5-touch'}).text.replace("\n", "")
```
